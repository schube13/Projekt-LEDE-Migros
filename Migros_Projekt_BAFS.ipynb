```python
df_C = pd.read_excel('Kaffee_preise.xlsx')
df_C.head()
```

```python

# Spalte 'Unnamed: 0' und Ã„hnliche bereinigen (optional, falls Indexspalten)
df_C = df_C.loc[:, ~df_C.columns.str.contains('^Unnamed')]

# Liste der Spalten ab Juni 2023 extrahieren
start_col = 'Jan 2023'
cols_to_keep = df_C.columns[df_C.columns.get_loc(start_col):]

# Neuen DataFrame mit nur diesen Spalten
df_C_filtered = df_C[cols_to_keep]

# Optional: Ausgabe zur Kontrolle
print(df_C_filtered.head())
```

```python
df_C_filtered.columns = pd.to_datetime(df_C_filtered.columns, format='%b %Y')

# Optional: Ausgabe zur Kontrolle
print(df_C_filtered.columns)
print(type(df_C_filtered.columns[0]))
```

```python
# Transponiere, damit jede Zeile ein Datum ist
df_C_transposed = df_C_filtered.transpose().reset_index()

# Umbenennen fÃ¼r Klarheit
df_C_transposed.columns = ['Date'] + [f'Kaffee_{i}' for i in range(1, len(df_C_transposed.columns))]

# Stelle sicher, dass Date wirklich datetime ist
df_C_transposed['Date'] = pd.to_datetime(df_C_transposed['Date'])
```

```python
# Nach Datum mergen
df_merged = pd.merge(df_clean, df_C_transposed, on='Date', how='outer').sort_values('Date')

```

```python
df_merged.plot(
    x='Date',
    y=['Nahrungsmittel', 'Energie'] + list(df_C_transposed.columns[1:]),
    figsize=(12, 6),
    title='Verlauf von Nahrungsmittel, Energie und Kaffee'
)
```

```python
df_export = df_merged[['Date', 'Nahrungsmittel', 'Energie'] + list(df_C_transposed.columns[1:])]
df_export['Date'] = df_export['Date'].dt.strftime('%Y-%m-%d')  # Optional: Datum als Text
df_export.to_csv('datawrapper_input.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Beispiel: Nur die ersten 15 Zeilen anzeigen (bei Bedarf anpassen)
df_table = df_export.head(15)

# Figure erstellen
fig, ax = plt.subplots(figsize=(12, 0.5 + 0.4 * len(df_table)))  # HÃ¶he dynamisch
ax.axis('off')  # keine Achsen anzeigen

# Tabelle einfÃ¼gen
table = plt.table(
    cellText=df_table.values,
    colLabels=df_table.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  # Spaltenbreite, ZeilenhÃ¶he anpassen

# SVG speichern
plt.savefig('energie_tabelle.svg', format='svg', bbox_inches='tight')
```
