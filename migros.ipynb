```python
df_Espresso[['Menge', 'Preis']] = df_Espresso.apply(
    lambda row: [row['Menge'] / 2, row['Preis'] / 2] if row['Menge'] == 2 else [row['Menge'], row['Preis']],
    axis=1,
    result_type='expand'
)
df_Espresso
```

```python
import pandas as pd

# Sicherstellen, dass 'Datum' als Datetime-Objekt interpretiert wird
for df in [df_Espresso, df_Zopf_500, df_Teig]:
    df['Datum'] = pd.to_datetime(df['Datum'])

# Daten zusammenführen
df_merged = pd.concat([df_Espresso, df_Zopf_500, df_Teig], axis=0, ignore_index=True)

# Optional: nach Datum sortieren
df_merged = df_merged.sort_values('Datum').reset_index(drop=True)

# Ausgabe anzeigen (optional)
print(df_merged.head())
```

```python
import matplotlib.pyplot as plt

# Datum sicherstellen
for df in [df_Espresso, df_Zopf_500, df_Teig]:
    df['Datum'] = pd.to_datetime(df['Datum'])

# Plot vorbereiten
plt.figure(figsize=(12, 6))

# Espresso
plt.plot(df_Espresso['Datum'], df_Espresso['Preis'], label='Espresso', color='brown', linewidth=2)

# Zopf
plt.plot(df_Zopf_500['Datum'], df_Zopf_500['Preis'], label='Zopf 500g', color='goldenrod', linewidth=2)

# Teig
plt.plot(df_Teig['Datum'], df_Teig['Preis'], label='Teig', color='olive', linewidth=2)

# Styling
plt.title('Preisverlauf von Espresso, Zopf und Teig')
plt.xlabel('Datum')
plt.ylabel('Preis (CHF)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Anzeigen
plt.show()
```
